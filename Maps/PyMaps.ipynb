```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('text',usetex=True)
plt.rc('font',family='serif')
plt.rcParams['xtick.labelsize']=13
plt.rcParams['ytick.labelsize']=13
plt.rcParams['axes.labelsize']=16
plt.rcParams['axes.titlesize']=16
```

```python
def Map(lon_low_left, lon_up_right, lat_low_left, lat_up_right):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_axes([0.1,0.1,0.8,0.8])
    m = Basemap(llcrnrlon=lon_low_left, llcrnrlat=lat_low_left, urcrnrlon=lon_up_right, urcrnrlat=lat_up_right,\
            rsphere=(6378137.00,6356752.3142),\
            resolution='l',projection='merc', epsg=5520)
    m.arcgisimage(service='NatGeo_World_Map', xpixels = 1000, verbose= True)
    meridians = np.linspace(lon_low_left, lon_up_right, num=5)
    m.drawmeridians(meridians, labels=[0,0,1,1], linewidth=1, fontsize=14)
    parallels = np.linspace(lat_low_left, lat_up_right, num=5)
    m.drawparallels(parallels, labels=[1,1,0,0], linewidth=1, fontsize=14)
    plt.show()
```

```python
Map(-22, -20.7, 63.7, 64.2)
```
